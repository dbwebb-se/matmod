## 1 Monte Carlo-metoden

I detta kursmoment så kommer vi gå igenom hur man kan uppskatta integraler.

Om $f(x)$ är en funktion som är icke-negativ på intervallet $[a,b]$ så kan vi beräkna arean $A$ av området mellan funktion och $x$-axeln med integralen

\begin{equation}
A = \int_a^b f(x) \; dx = F(b) - F(a)
\end{equation}

där $F(x)$ uppfyller $F'(x) = f(x)$.

Dock så är det inte alltid möjligt att finna $F(x)$ så att vi kan beräkna integralen.

Monte Carlo-metoden är en probabilistisk metod som uppskattar värdet på en integral. Låt $c$ vara det största funktionsvärdet av $f(x)$ på intervallet $[a,b]$. Då gäller det att

\begin{equation}
A = \int_a^b f(x) \; dx \leq (b-a)\cdot c
\end{equation}
dvs arean $A$ är mindre än arean av den rektangel som innesluter omsluter området. Följande kodexempel illusterar hur en rektangel innsluter en kurva i intervallet $[0,4]$.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

a = 0
b = 4
x = np.linspace(a,b,1000)
y = np.absolute(x*np.exp(-np.sin(x)**2))

plt.plot(x, y)

currentAxis = plt.gca()
currentAxis.add_patch(Rectangle((0, 0), 4, 3.22, facecolor="blue", alpha = 0.1))
plt.savefig('area.png')
plt.grid(True)
plt.show()
```

Monte Carlo-metoden fungerar på följande sätt. Betrakta händelsen $U$ att en slumpvis vald punkt inuti rektangeln ligger under grafen $y = f(x)$.

Generera $n$ punkter slumpvis som ligger inuti rektangeln.
Beräkna antalet punkter som ligger under grafen $y = f(x)$. Beteckna detta tal med $m$.
Då $n$ växer så gäller det att $m/n$ går mot $Pr(U)$

I figuren nedan är 200 punkter slumpvis plottade. De blå punkterna ligger under grafen och de oranga ligger över grafen.

```python
from IPython.display import Image
Image("montecarlo.png")
```

När vi vet sannolikheten för att en slumpvis vald punkt ligger under grafen så kan uppskatta integralen enligt

\begin{equation}
A \approx \Pr(U)\cdot (b-a)\cdot c
\end{equation}

## 2 Centrala gränsvärdessatsen

Centrala gränsvärdessatsen säger att n oberoende slumpvariabler med samma fördelning är ungefär normalfördelad om $n$ är tillräckligt stort. Om $X_1, X_2, X_n$ är oberoende slumpvariabler så gäller det

\begin{equation}
Y = X_1 + X_2 + \ldots + X_n
\end{equation}

är approximativt normalfördelad.

Exempel på försök med oberoende slumpvariabler är att kasta ett stort fixerat antal tärningar och notera summan av tärningarna. Om man genomför detta slumpförsök många gånger så kommer summan av tärningarna vara normalfördelad.
